The test heuristic file is the output of running e-SNLI model against the HANS dataset. There are 3 fallable syntactic heuristics tested by the HANS dataset: lexical overlap, subsequence and constituent.

For each heuristic, I selected 3200 samples to run the test on e-SNLI model. Within 3200 samples for each heuristic, 1600 samples are entailment and 1600 samples are non-entailment

This notebook will analyze the result after running the model on the sample HANS dataset.

```python
import pandas as pd
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_csv('HANS_output.csv')
```

```python
lexical_overlap_df = df[df['heuristic'] == 'lexical_overlap']
subsequence_df = df[df['heuristic'] == 'subsequence']
constituent_df = df[df['heuristic'] == 'constituent']
print("Number of lexical_overlap tests: ", len(lexical_overlap_df))
print("Number of subsequence tests: ", len(subsequence_df))
print("Number of constituent tests: ", len(constituent_df))
```

```python
# count the number of tests where predicted label is entailment but the gold_label is not entailment
lexical_overlap_entailment = lexical_overlap_df[(lexical_overlap_df['gold_label'] == 'neutral') & (lexical_overlap_df['pred_label'] == 'entailment')]
print("Number of non-entailment examples that are misclassified as entailment (lexical overlap) ", \
      len(lexical_overlap_entailment))
print('Error rate: ', len(lexical_overlap_entailment)/1600 * 100, '%')
print('Accuracy: ', 100 - len(lexical_overlap_entailment)/1600 * 100, '%')
```

```python
lexical_overlap_non_entailment = lexical_overlap_df[(lexical_overlap_df['gold_label'] == 'entailment') & (lexical_overlap_df['pred_label'] != 'entailment')]
print("Number of entailment examples that are misclassified as non-entailment (lexical overlap) ", \
        len(lexical_overlap_non_entailment))
print('Error rate: ', len(lexical_overlap_non_entailment)/1600 * 100, '%')
print('Accuracy: ', 100 - len(lexical_overlap_non_entailment)/1600 * 100, '%')
```

```python
subsequence_entailment = subsequence_df[(subsequence_df['pred_label'] == 'entailment') & (subsequence_df['gold_label'] == 'neutral')]
print("Number of non-entailment examples that are misclassified as entailment (subsequence) ", \
        len(subsequence_entailment))
print('Error rate: ', len(subsequence_entailment)/1600 * 100, '%')
print('Accuracy: ', 100 - len(subsequence_entailment)/1600 * 100, '%')
```

```python
subsequence_non_entailment = subsequence_df[(subsequence_df['pred_label'] != 'entailment') & (subsequence_df['gold_label'] == 'entailment')]
print("Number of entailment examples that are misclassified as non-entailment (subsequence) ", \
        len(subsequence_non_entailment))
print('Error rate: ', len(subsequence_non_entailment)/1600 * 100, '%')
print('Accuracy: ', 100 - len(subsequence_non_entailment)/1600 * 100, '%')
```

```python
constituent_entailment = constituent_df[(constituent_df['pred_label'] == 'entailment') & (constituent_df['gold_label'] == 'neutral')]
print("Number of non-entailment examples that are misclassified as entailment (constituent) ", \
        len(constituent_entailment))
print('Error rate: ', len(constituent_entailment)/1600 * 100, '%')
print('Accuracy: ', 100 - len(constituent_entailment)/1600 * 100, '%')
```

```python
constituent_non_entailment = constituent_df[(constituent_df['pred_label'] != 'entailment') & (constituent_df['gold_label'] == 'entailment')]
print("Number of entailment examples that are misclassified as non-entailment (constituent) ", \
        len(constituent_non_entailment))    
print('Error rate: ', len(constituent_non_entailment)/1600 * 100, '%')
print('Accuracy: ', 100 - len(constituent_non_entailment)/1600 * 100, '%')
```
